```python
import random

# Configuración inicial
print("Juego sencillo de BLACK TIGER 'Adivina el Número' ")

jugar_nuevamente = "sí"  # Iniciamos el bucle principal

while jugar_nuevamente.lower() in ["sí", "si", "s", "yes", "y"]:  # Depende de lo que el usuario escoja
    # Generar número aleatorio entre el uno y 10 porque asi se puso el random y el intentos es para demarcar cuantas oportunidades tiene
    numero_secreto = random.randint(1, 10)
    intentos = 3

    print("\nTienes que adivinar un número entre 1 y 10. ¡Tienes 3 intentos!")

    # Bucle del juego hasta completar los intentos
    while intentos > 0:
        try:
            numero_usuario = int(input("Tu intento: "))

            if numero_usuario == numero_secreto:
                print(f"¡Correcto! El número era {numero_secreto}. ")
                break
            else:
                intentos -= 1
                if intentos > 0:
                    print(f" Incorrecto. Te quedan {intentos} intentos.")
                else:
                    print(f"¡Perdiste! El número era {numero_secreto}. ")

        except ValueError:        #Evita que el programa se bloquee si el usuario ingresa algo invalido como letras.
            print("Error: Debes ingresar un número entero.")

    # Preguntar si quiere jugar de nuevo
    jugar_nuevamente = input("\n¿Quieres jugar otra vez? (si/no): ")

print("\n¡Gracias por jugar! Hasta la próxima. BLACK TIGER ")
```
