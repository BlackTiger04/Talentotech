```python
import itertools
import string
from collections import Counter

# Diccionario de palabras en español (PARA EL EJEMPLO, LO MEJOR ES USAR DICIONARIOS COMPLETOS.)
SPANISH_DICTIONARY = [
    'DE', 'LA', 'QUE', 'EL', 'EN', 'Y', 'A', 'LOS', 'DEL', 'SE',
    'LAS', 'UN', 'POR', 'CON', 'NO', 'UNA', 'SU', 'PARA', 'ES',
    'AL', 'LO', 'COMO', 'MÁS', 'PERO', 'SUS', 'LE', 'HA', 'ME',
    'SI', 'SIN','SOL','SAL', 'SOBRE', 'ESTE','HOLA','DIA', 'YA', 'ENTRE', 'CUANDO', 'TODO',
    'ESTA', 'SER', 'SON', 'DOS', 'TAMBIÉN', 'FUÉ', 'HABIA', 'ERA',
    'MUY', 'AÑOS', 'HASTA', 'DESDE', 'ESTÁN', 'MI', 'PORQUE', 'QUÉ',
    'SÓLO', 'HAN', 'YO', 'HAY', 'VEZ', 'PUEDE', 'TODOS', 'ASÍ',
    'NOS', 'NI', 'PARTE', 'TENÍA', 'ÉL', 'ELLA', 'ESTAR', 'ESTÁ',
    'AQUÍ', 'HACER', 'OTRO', 'ERA', 'ERAN', 'AÑO', 'MES', 'DÍA',
    'TIEMPO', 'MUNDO', 'VIDA', 'PERSONA', 'FORMA', 'CASO', 'HOMBRE',
    'SEÑOR', 'MUJER', 'CIUDAD', 'PAÍS', 'LUGAR', 'TRABAJO', 'GRUPO',
    'NOCHE', 'PUNTO', 'ESTADO', 'NÚMERO', 'FAMILIA', 'AGUA', 'LUZ'
]

def vigenere_encrypt(plain_text, key):
    encrypted_text = []
    key_length = len(key)

    for i, char in enumerate(plain_text):
        if char.isalpha():
            char_upper = char.upper()
            plain_num = ord(char_upper) - ord('A')
            key_num = ord(key[i % key_length].upper()) - ord('A')
            encrypted_num = (plain_num + key_num) % 26
            encrypted_char = chr(encrypted_num + ord('A'))
            encrypted_text.append(encrypted_char)
        else:
            encrypted_text.append(char)

    return ''.join(encrypted_text)

def vigenere_decrypt(encrypted_text, key):
    decrypted_text = []
    key_length = len(key)

    for i, char in enumerate(encrypted_text):
        if char.isalpha():
            char_upper = char.upper()
            encrypted_num = ord(char_upper) - ord('A')
            key_num = ord(key[i % key_length].upper()) - ord('A')
            plain_num = (encrypted_num - key_num) % 26
            plain_char = chr(plain_num + ord('A'))
            decrypted_text.append(plain_char)
        else:
            decrypted_text.append(char)

    return ''.join(decrypted_text)

def calculate_meaningfulness(text, dictionary):
    """Calcula un puntaje de legibilidad basado en palabras del diccionario"""
    words = text.split()
    if not words:
        return 0

    # Contar palabras reconocidas (ignorando mayúsculas/minúsculas)
    recognized = sum(1 for word in words if word.upper() in dictionary)
    return recognized / len(words)  # Porcentaje de palabras reconocidas

def smart_brute_force(encrypted_text, max_key_length=4, dictionary=SPANISH_DICTIONARY):
    """Ataque por fuerza bruta inteligente con puntuación de resultados"""
    results = []
    alphabet = 'ABCDEFGHIJKLMNÑOPQRSTUVWXYZ'

    print(f"\nIniciando ataque inteligente (claves hasta {max_key_length} caracteres)...")

    for key_length in range(1, max_key_length + 1):
        print(f"Probando claves de longitud {key_length}...")

        for possible_key in itertools.product(alphabet, repeat=key_length):
            key = ''.join(possible_key)
            decrypted = vigenere_decrypt(encrypted_text, key)
            score = calculate_meaningfulness(decrypted, dictionary)

            if score > 0.3:  # Lo que esta en el diccinario mínimo de palabras reconocidas
                results.append((score, key, decrypted))

    # Ordenar resultados por puntaje dando por entendido las palabras del dicionario (mejores primero)
    results.sort(reverse=True, key=lambda x: x[0])

    if results:
        print("\nResultados más probables:")
        for i, (score, key, decrypted) in enumerate(results[:3]):
            print(f"\nOpción {i+1}:")
            print(f"Clave: {key}")
            print(f"Puntuación: {score:.2f}")
            print(f"Mensaje: {decrypted}")

        # Seleccionar automáticamente el mejor resultado
        best_score, best_key, best_decrypted = results[0]
        print(f"\nSeleccionado automáticamente el mejor resultado:")
        print(f"Clave encontrada: {best_key}")
        print(f"Mensaje descifrado: {best_decrypted}")

        return best_key, best_decrypted
    else:
        print("\nNo se encontraron resultados satisfactorios.")
        print("Prueba aumentando la longitud máxima de clave o verifica el mensaje.")
        return None, None

def main():
    while True:
        print("\n  Menú del Cifrado de Vigenère ")
        print("1. Cifrar un mensaje")
        print("2. Descifrar un mensaje (con clave)")
        print("3. Ataque inteligente por fuerza bruta (sin clave)")
        print("4. Salir")

        option = input("Seleccione una opción (1/2/3/4): ").strip()

        if option == '1':
            print("\n--- Cifrado ---")
            mensaje = input("Ingrese el mensaje a cifrar: ")
            clave = input("Ingrese la clave de cifrado: ")

            if not all(c.isalpha() for c in clave):
                print("Error: La clave solo debe contener letras.")
                continue

            mensaje_cifrado = vigenere_encrypt(mensaje, clave)
            print(f"\nMensaje original: {mensaje}")
            print(f"Clave usada: {clave}")
            print(f"Mensaje cifrado: {mensaje_cifrado}")

        elif option == '2':
            print("\n--- Descifrado ---")
            mensaje_cifrado = input("Ingrese el mensaje cifrado: ")
            clave = input("Ingrese la clave de descifrado: ")

            if not all(c.isalpha() for c in clave):
                print("Error: La clave solo debe contener letras.")
                continue

            mensaje_descifrado = vigenere_decrypt(mensaje_cifrado, clave)
            print(f"\nMensaje cifrado: {mensaje_cifrado}")
            print(f"Clave usada: {clave}")
            print(f"Mensaje descifrado: {mensaje_descifrado}")

        elif option == '3':
            print("\n  Ataque ed fuerza con diccionario  ")
            mensaje_cifrado = input("Ingrese el mensaje cifrado: ")
            max_longitud = input("Longitud máxima de clave a probar (1-4): ").strip()

            try:
                max_longitud = int(max_longitud)
                if max_longitud < 1 or max_longitud > 4:
                    print("Usando valor por defecto 4 (puede ser lento)")
                    max_longitud = 4
            except ValueError:
                print("Valor inválido. Usando longitud máxima 4.")
                max_longitud = 4

            clave, mensaje = smart_brute_force(mensaje_cifrado, max_key_length=max_longitud)

            if clave:
                print("\n¡Descifrado exitoso!")
                print(f"Clave probable: {clave}")
                print(f"Mensaje descifrado: {mensaje}")

        elif option == '4':
            print("Saliendo del programa...")
            break

        else:
            print("Opción no válida. Por favor ingrese 1, 2, 3 o 4.")

if __name__ == "__main__":
    main()
```

```python
# Función principal con lambada y numeros primos hasta 100
encontrar_primos = lambda cantidad: [
    x for x in range(2, 1000)
    if len([i for i in range(1, x+1) if x % i == 0]) == 2
][:cantidad]

primos = encontrar_primos(100)

print("Los primeros 100 números primos son:")
for i, primo in enumerate(primos, 1):
    print(f"{i:3}: {primo}", end=' | ')      # print(f"{i:3}: {primo}", end=' | ') De la ia Para que aparezcan bien organizados.
    if i % 5 == 0:
        print()
```
