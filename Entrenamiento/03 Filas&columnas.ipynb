```python
# Lista con información de 10 estudiantes
estudiantes = [
    {"nombre": "Lorena", "apellido": "Paz", "asignatura": "Matemáticas", "nota_final": 8.5},
    {"nombre": "Pedro", "apellido": "Arzuaga", "asignatura": "Español", "nota_final": 5.9},
    {"nombre": "Hector", "apellido": "Castro", "asignatura": "Historia", "nota_final": 9.2},
    {"nombre": "Carlos", "apellido": "Bolaños", "asignatura": "Física", "nota_final": 6.0},
    {"nombre": "Elena", "apellido": "Daza", "asignatura": "Química", "nota_final": 4.5},
    {"nombre": "Miguel", "apellido": "Salazar", "asignatura": "Ingles", "nota_final": 7.3},
    {"nombre": "Isabel", "apellido": "Moreno", "asignatura": "Arte", "nota_final": 10.0},
    {"nombre": "Diego", "apellido": "Vargas", "asignatura": "Biología", "nota_final": 3.8},
    {"nombre": "Karla", "apellido": "Torres", "asignatura": "Geografía", "nota_final": 6.7},
    {"nombre": "Javier", "apellido": "Herrera", "asignatura": "Informatica", "nota_final": 2.9},
]

# Función para mostrar estudiantes según criterio
def mostrar_estudiantes(filtro):
    print(f"\n{'Nombre':<10} {'Apellido':<12} {'Asignatura':<20} {'Nota Final':<10}")
    print("-" * 55)

    for est in estudiantes:
        nota = est["nota_final"]
        if (filtro == "todos" or
            (filtro == "aprobados" and nota >= 6.0) or
            (filtro == "reprobados" and nota < 6.0)):
            print(f"{est['nombre']:<10} {est['apellido']:<12} {est['asignatura']:<20} {nota:<10}")

# Menú de opciones
def menu():
    print("\n=== MENÚ DE OPCIONES ===")
    print("1. Mostrar todos los estudiantes")
    print("2. Mostrar solo aprobados")
    print("3. Mostrar solo reprobados")
    print("4. Salir")

# Programa principal
while True:
    menu()
    opcion = input("Selecciona una opción (1-4): ")

    if opcion == "1":
        mostrar_estudiantes("todos los estudiantes")
    elif opcion == "2":
        mostrar_estudiantes("aprobados")
    elif opcion == "3":
        mostrar_estudiantes("reprobados")
    elif opcion == "4":
        print("Programa finalizado.")
        break
    else:
        print("Opción no válida. Intenta de nuevo.")

```
