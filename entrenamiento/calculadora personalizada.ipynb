```python
import math

def mostrar_menu():
    print("\n  CALCULADORA  ")
    print("1. Para sumar")
    print("2. Para Restar ")
    print("3. Para Multiplicar ")
    print("4. Para Dividir ")
    print("5. Potenciación ")
    print("6. Logaritmo solo a base 10")
    print("7. Seno (sin)")
    print("8. Coseno (cos)")
    print("9. Tangente (tan)")
    print("0. Salir")

def pedir_dos_numeros():
    num1 = float(input("Ingrese el primer número: "))
    num2 = float(input("Ingrese el segundo número: "))
    return num1, num2

def pedir_un_numero():
    num = float(input("Ingrese un número: "))
    return num

def calculadora():
    while True:
        mostrar_menu()
        opcion = input("Seleccione una opción (0-9): ")

        if opcion == "0":
            print("Saliendo de la calculadora. ¡Hasta luego!")
            break

        elif opcion == "1":
            numero1, numero2 = pedir_dos_numeros()
            print("Resultado suma:", numero1 + numero2)

        elif opcion == "2":
            numero1, numero2 = pedir_dos_numeros()
            print("Resultado resta:", numero1 - numero2)

        elif opcion == "3":
            numero1, numero2 = pedir_dos_numeros()
            print("Resultado multiplicacion:", numero1 * numero2)

        elif opcion == "4":
            numero1, numero2 = pedir_dos_numeros()
            if numero2 != 0:
                print("Resultado division:", numero1 / numero2)
            else:
                print("Recuerda buen amigo, no se puede dividir por cero :P ")

        elif opcion == "5":
            numero1, numero2 = pedir_dos_numeros()
            print("Resultado potenciado:", numero1 ** numero2)

        elif opcion == "6":
            numero1 = pedir_un_numero()
            if numero1 > 0:
                print("Resultado logarigmo es :", math.log10(numero1))
            else:
                print("Error: el logaritmo solo está definido para números positivos.")

        elif opcion == "7":
            numero1 = pedir_un_numero()
            print("Resultado en radianes del seno: ", math.sin(math.radians(numero1)))  # Convierte a radianes

        elif opcion == "8":
            numero1 = pedir_un_numero()
            print("Resultado en radianes del coseno:", math.cos(math.radians(numero1)))  # Convierte a radianes

        elif opcion == "9":
            numero1 = pedir_un_numero()
            print("Resultado en radianes del tangente:", math.tan(math.radians(numero1)))  # Convierte a radianes

        else:
            print("Opción inválida no meta otra cosa que no sea numero. Por favor, seleccione una opción del 0 al 9.")

# Ejecutar la calculadora
calculadora()
```
